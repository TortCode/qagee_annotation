```python
import json

with open('event_role_KAIROS.json', 'r') as inf:
    event_roles = json.load(inf)
```

```python
for key, value in event_roles.items():
    template = value['template']
    roles = value['roles']
    for i, role in enumerate(roles, start=1):
        template = template.replace(f"<arg{i}>", f"[{role}]", 1)
    event_roles[key]['template'] = template

event_roles
```

```python
with open('event_role_formatted.json', 'w') as outf:
    json.dump(event_roles, outf, indent=2)
```
